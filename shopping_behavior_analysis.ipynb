```python

labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']

df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

print(df[['age', 'age_group']].head(10))
```

```python
frequency_mapping = {
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Fortnightly': 14,  
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 Months': 90,
    'Annually': 365
}


df['frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)


print(df[['frequency_of_purchases', 'frequency_days']].head())
```

```python

spending_by_age = df.groupby('age_group', observed=True)['purchase_amount_usd'].sum().sort_values(ascending=False)
print(spending_by_age)
```

```python
sub_vs_non = df.groupby('subscription_status')['purchase_amount_usd'].mean()
print(sub_vs_non)
```

```python
top_states = df.groupby('location')['purchase_amount_usd'].sum().nlargest(5)
print("\nTop 5 States by Revenue:\n", top_states)
```

```python
# 1. Most Popular Categories
category_counts = df['category'].value_counts()
print("Orders per Category:\n", category_counts)

# 2. Category Popularity by Season
seasonal_trends = df.groupby(['season', 'category']).size().unstack()
print("\nSeasonal Category Trends:\n", seasonal_trends)
```

```python
# 1. Average Spend: Subscriber vs Non-Subscriber
sub_spend = df.groupby('subscription_status')['purchase_amount_usd'].mean()
print("Average Spend by Subscription Status:\n", sub_spend)

# 2. Promo Code Impact
promo_impact = df.groupby('promo_code_used')['purchase_amount_usd'].mean()
print("\nAverage Spend: Promo Code vs No Promo:\n", promo_impact)
```

```python
# 1. Most Popular Payment Method
payment_methods = df['payment_method'].value_counts()
print("Payment Method Preferences:\n", payment_methods)

# 2. Shipping Type vs. Review Rating
# Does faster shipping lead to better reviews?
shipping_reviews = df.groupby('shipping_type')['review_rating'].mean().sort_values(ascending=False)
print("\nAverage Review Rating by Shipping Type:\n", shipping_reviews)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

age_spending = df.groupby('age_group', observed=True)['purchase_amount_usd'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=age_spending.index, y=age_spending.values, palette='viridis')

plt.title('Total Revenue by Age Group')
plt.ylabel('Total Spend (USD)')
plt.xlabel('Age Group')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Promo Code vs. Loyalty (Basket Analysis)
# Do promo code users have more 'previous purchases'?
promo_loyalty = df.groupby('promo_code_used')['previous_purchases'].mean()
print("Average Previous Purchases (Promo vs No):")
print(promo_loyalty)

# 2. The "Unhappy State" Finder
# Which states have the lowest average review ratings? 
unhappy_states = df.groupby('location')['review_rating'].mean().nsmallest(5)
print("\nStates with Lowest Review Ratings:")
print(unhappy_states)

# 3. Age vs. Payment Method
# Do Seniors use different methods than Young Adults?
age_payment = pd.crosstab(df['age_group'], df['payment_method'], normalize='index') * 100
print("\nPayment Method % by Age Group:")
print(age_payment)

# 4. Seasonal Color Trends
# Does the top color change in Summer vs Winter?
seasonal_colors = df.groupby('season')['color'].agg(lambda x: x.value_counts().index[0])
print("\nMost Popular Color per Season:")
print(seasonal_colors)

# 5. Shipping Luxury
# Which category has the most 'Express' shipping orders?
express_shipping = df[df['shipping_type'] == 'Express']['category'].value_counts()
print("\nExpress Shipping Orders by Category:")
print(express_shipping)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# 1. Promo Code vs. Loyalty (Bar Plot)
plt.figure()
sns.barplot(x=promo_loyalty.index, y=promo_loyalty.values, palette='RdYlGn')
plt.title('Does Promo Code Use Indicate Higher Customer Loyalty?')
plt.ylabel('Average Previous Purchases')
plt.show()

# 2. The "Unhappy State" Finder (Horizontal Bar Chart)
plt.figure()
sns.barplot(x=unhappy_states.values, y=unhappy_states.index, palette='Reds_r')
plt.title('Top 5 States with Lowest Review Ratings')
plt.xlabel('Average Rating (1-5)')
plt.xlim(0, 5) 
plt.show()

# 3. Age vs. Payment Method (Heatmap)
# This is a very "Pro" way to show complex relationships
plt.figure(figsize=(12, 6))
sns.heatmap(age_payment, annot=True, cmap='YlGnBu', fmt='.1f')
plt.title('Payment Method Preference by Age Group (%)')
plt.show()

# 4. Seasonal Color Trends (Count Plot with Hue)
# We will show the top colors across seasons
plt.figure()
# We'll filter for the top 5 most common colors to keep the chart clean
top_colors = df['color'].value_counts().nlargest(5).index
filtered_df = df[df['color'].isin(top_colors)]

sns.countplot(data=filtered_df, x='season', hue='color', palette='viridis')
plt.title('Color Popularity Across Seasons')
plt.show()

# 5. Shipping Luxury (Donut Chart)
plt.figure()
plt.pie(express_shipping, labels=express_shipping.index, autopct='%1.1f%%', 
  startangle =140, pctdistance=0.85, colors=sns.color_palette('pastel'))

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Which Categories Use Express Shipping Most?')
plt.show() 
```
